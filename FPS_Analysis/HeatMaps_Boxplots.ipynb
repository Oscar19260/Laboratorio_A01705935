```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
import numpy as np; np.random.seed(0)
from matplotlib import cm
```

```python
df = pd.read_csv('dataset/Fortnite_players_stats.csv', index_col = 0)
new_df = df[['Solo minutesPlayed', 'Duos minutesPlayed', 'Trios minutesPlayed','Squads minutesPlayed', 'LTM minutesPlayed']].describe()
new_df
```

```python
plt.plot(df['Solo minutesPlayed'], df['Solo kills'], '*')
plt.xlabel("Solo minutes played")
plt.ylabel("Solo kills")
plt.show()
```

```python
plt.plot(df['Duos minutesPlayed'], df['Duos kills'], '*')
plt.xlabel("Duos minutes played")
plt.ylabel("Duos kills")
plt.show()
```

```python
plt.plot(df['Trios minutesPlayed'], df['Trios kills'], '*')
plt.xlabel("Trios minutes played")
plt.ylabel("Trios kills")
plt.show()
```

```python
plt.plot(df['Squads minutesPlayed'], df['Squads kills'], '*')
plt.xlabel("Squads minutes played")
plt.ylabel("Squads kills")
plt.show()
```

```python
plt.plot(df['LTM minutesPlayed'], df['LTM kills'], '*')
plt.xlabel("LTM minutes played")
plt.ylabel("LTM kills")
plt.show()
```

```python
%matplotlib inline
df[['Solo minutesPlayed', 'Duos minutesPlayed']].plot.hist(bins=7, alpha=0.5)
plt.show ()
```

```python
df[[ 'Trios minutesPlayed', 'Squads minutesPlayed', 'LTM minutesPlayed']].plot.hist(bins=7, alpha=0.5)
plt.show ()
```

```python
df = df[['Solo minutesPlayed', 'Duos minutesPlayed', 'Trios minutesPlayed','Squads minutesPlayed', 'LTM minutesPlayed']]
colormap = plt.cm.viridis
plt.figure(figsize=(10,10))
plt.title('Pearson Correlation of Features', y=1.05, size=18)
sb.heatmap(df.astype(float).corr(), linewidths=0.1, vmax=1.0, square=True, 
           cmap=colormap, linecolor='white', annot=True);
```
